# 1.Лемматизация и стемминг на любом тексте

Стемминг – отрезает «лишнее» от корня слов.

Лемматизация – приводит слово к его канонической форме.


---



Подготовка текста

```python
morph = pymorphy2.MorphAnalyzer()

text = """Меня зовут Кира Йошикаге. Мне 33 года. Мой дом находится в северо-восточной части Морио,
в районе поместий. Работаю в офисе сети магазинов и домой возвращаюсь, самое позднее, в восемь вечера.
Не курю, выпиваю изредка. Ложусь спать в 11 вечера и убеждаюсь, что получаю ровно восемь часов сна,
несмотря ни на что. Перед сном я пью тёплое молоко, а также минут двадцать уделяю разминке, поэтому
до утра сплю без особых проблем. Утром я просыпаюсь, не чувствуя ни усталости, ни стресса, словно младенец.
На медосмотре мне сказали, что никаких проблем нет. Я пытаюсь донести, что я обычный человек, который
хочет жить спокойной жизнью. Я не забиваю себе голову проблемами вроде побед или поражений, и не обзавожусь
врагами, из-за которых не мог бы уснуть. Я знаю наверняка: в таком способе взаимодействия с обществом
и кроется счастье. Хотя, если бы мне пришлось сражаться, я бы никому не проиграл."""


text = text.lower()
text = re.sub(r'[^\w\s]', '', text)
words = text.split()
```

Лемматизация:

```python
lemmatized_words = []
for word in words:
    lemma = morph.parse(word)[0].normal_form
    lemmatized_words.append(lemma)

lemmatized_text = " ".join(lemmatized_words)

print("Лемматизированный текст:")
print(lemmatized_text)
```

Стемминг:

```python
stemmed_words = []
for word in words:
    stem = morph.parse(word)[0].normal_form[:5]
    stemmed_words.append(stem)

stemmed_text = " ".join(stemmed_words)

print("\nСтиммингованный текст:")
print(stemmed_text)
```

# 2 и 3.Написать функцию токенизации и векторизации всех символов из ASCII

Токенизация – это процесс разделения предложений на слова-компоненты.

Векторизация – это подход к преобразованию входных данных из их исходного формата в векторы действительных чисел, которые понятны моделям машинного обучения.

Регулярное выражение – это последовательность символов, которая определяет шаблон поиска.

---



```python
def tokenize(text):
    pattern = r'\w+'
    tokens = re.findall(pattern, text)
    return tokens

def vectorize(tokens):
    vectorized = []
    for word in tokens:
        word_ascii = []
        for char in word:
            word_ascii.append(ord(char))
        vectorized.append(word_ascii)
    return vectorized


text = """Меня зовут Кира Йошикаге. Мне 33 года. Мой дом находится в северо-восточной части Морио,
в районе поместий. Работаю в офисе сети магазинов и домой возвращаюсь, самое позднее, в восемь вечера.
Не курю, выпиваю изредка. Ложусь спать в 11 вечера и убеждаюсь, что получаю ровно восемь часов сна,
несмотря ни на что. Перед сном я пью тёплое молоко, а также минут двадцать уделяю разминке, поэтому
до утра сплю без особых проблем. Утром я просыпаюсь, не чувствуя ни усталости, ни стресса, словно младенец.
На медосмотре мне сказали, что никаких проблем нет. Я пытаюсь донести, что я обычный человек, который
хочет жить спокойной жизнью. Я не забиваю себе голову проблемами вроде побед или поражений, и не обзавожусь
врагами, из-за которых не мог бы уснуть. Я знаю наверняка: в таком способе взаимодействия с обществом
и кроется счастье. Хотя, если бы мне пришлось сражаться, я бы никому не проиграл."""


tokenized_text= tokenize(text)
print(tokenized_text[:5])

vectorized_text = vectorize(tokenized_text)
print(vectorized_text[:5])
```
